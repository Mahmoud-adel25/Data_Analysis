```python
import pandas as pd
import seaborn as sns
import statistics as st
import numpy as np
import matplotlib.pyplot as mt
df = pd.read_csv("C:/Users/mahmo/OneDrive/Documents/COLLEGE/DATA ANALYSIS PROJECT/Diabetes Dataset.csv")
```

# Correlation Matrix 

```python
corr_matrix=df.corr()
mt.figure(figsize=(8, 6))
mt.title('Correlation Matrix')
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
mt.show()
```

# Scatter Plot 

```python
mt.scatter(df['BloodPressure'],df['BMI'])
mt.xlabel('BloodPressure')
mt.ylabel('BMI')
x = [0, 180]  
y = [0, 60]   
mt.plot(x, y, color='red', linestyle='-')
mt.title('Scatter Plot between BloodPressure and BMI')
mt.show()
```

# DISTRIBUTION PLOT

```python
hist, bins = np.histogram(df['Pregnancies'], density=True)

bin_centers = (bins[:-1] + bins[1:]) / 2
mt.hist(df['Pregnancies'], bins=bins, edgecolor='black', density=True)

mt.plot(np.concatenate(([0], bin_centers)), np.concatenate(([0], hist)), color='red') # concatenate 34an yebda2 mn zero fe el x-axis
mt.scatter(bin_centers, hist, color='red', marker='o')
mt.xlabel('Pregnancies')
mt.ylabel('Frequency')
mt.title('Histogram and Frequency Polygon of Pregnancies')
mt.show()
```

```python
hist, bins = np.histogram(df['SkinThickness'], density=True)
bin_centers = (bins[:-1] + bins[1:]) / 2
mt.hist(df['SkinThickness'], bins=bins, edgecolor='black', density=True)
mt.plot(np.concatenate(([0], bin_centers)), np.concatenate(([0], hist)), color='red') # concatenate 34an yebda2 mn zero fe el x-axis
mt.scatter(bin_centers, hist, color='red', marker='o')
mt.xlabel('SkinThickness')
mt.ylabel('Frequency')
mt.title('Histogram and Frequency Polygon of SkinThickness')
mt.show()
```

```python
hist, bins = np.histogram(df['Age'], density=True)
bin_centers = (bins[:-1] + bins[1:]) / 2
mt.hist(df['Age'], bins=bins, edgecolor='black', density=True)
mt.plot(np.concatenate(([0], bin_centers)), np.concatenate(([0], hist)), color='red') # concatenate 34an yebda2 mn zero fe el x-axis
mt.scatter(bin_centers, hist, color='red', marker='o')
mt.xlabel('Age')
mt.ylabel('Frequency')
mt.title('Histogram and Frequency Polygon of Age')
mt.show()
```
