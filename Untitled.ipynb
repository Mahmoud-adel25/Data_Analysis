# Handle Zeros

```python
# Replace zeros with NaN to indicate missing/invalid values
columns_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[columns_with_zeros] = df[columns_with_zeros].replace(0, np.nan)
```

```python
df[columns_with_zeros] = df[columns_with_zeros].fillna(np.nan)
df[columns_with_zeros] = df[columns_with_zeros].astype(float)
```

```python
# Initialize the KNN Imputer
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform')  # Use 5 nearest neighbors
```

```python
# Fit and transform the data to impute missing values
df[columns_with_zeros] = knn_imputer.fit_transform(df[columns_with_zeros])

# Verify the imputed values
print(df.describe())
```

# Logistic Regression

```python
# Step 1: Impute missing values using KNN on the entire DataFrame
imputer = KNNImputer(n_neighbors=5)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Separate the imputed data back into features (X_imputed) and labels (y_imputed)
X_imputed = df_imputed.drop('Outcome', axis=1).values  # Features
y_imputed = df_imputed['Outcome'].values  # Target variable

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.2, random_state=42, stratify=y_imputed)

# Step 3: Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Train the Logistic Regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Step 5: Predict on the test set
y_pred = model.predict(X_test_scaled)

# Step 6: Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
mt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
mt.title('Confusion Matrix')
mt.xlabel('Predicted')
mt.ylabel('Actual')
mt.show()
```

```python
# Step 7: ROC Curve
y_prob = model.predict_proba(X_test_scaled)[:, 1]  # Use scaled data for ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
mt.figure(figsize=(8, 6))
mt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})')
mt.plot([0, 1], [0, 1], 'r--', label='Random Guess')
mt.title('Receiver Operating Characteristic (ROC) Curve')
mt.xlabel('False Positive Rate')
mt.ylabel('True Positive Rate')
mt.legend(loc='lower right')
mt.show()

# Print AUC Score
print(f"ROC AUC Score: {roc_auc:.2f}")
```

# Handling Outliers

## SkinThickness

```python
# 1. Identifying Outliers Using IQR Method
Q1 = df['SkinThickness'].quantile(0.25)
Q3 = df['SkinThickness'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
```

```python
# 2. Removing Outliers Using IQR
df_no_outliers_iqr = df[(df['SkinThickness'] >= lower_bound) & (df['SkinThickness'] <= upper_bound)]
print("Data after removing outliers using IQR:")
print(df_no_outliers_iqr['SkinThickness'].head())
```

```python


# 6. Capping Outliers Using Percentile-based Winsorization
lower_cap = df['SkinThickness'].quantile(0.05)
upper_cap = df['SkinThickness'].quantile(0.95)
df['SkinThickness' + '_capped'] = np.clip(df['SkinThickness'], lower_cap, upper_cap)
print(f"  Data after capping outliers for {'SkinThickness'}:")
print(df[['SkinThickness', 'SkinThickness' + '_capped']].head())

```

```python
mt.figure(figsize=(12, 6))
mt.subplot(2, 2, 2)
mt.boxplot(df_no_outliers_iqr['SkinThickness'],vert=False)
mt.title(f'After IQR Handling Boxplot')
mt.tight_layout()
```

# Feature Selection


```python
corr_matrix = df.corr()

# Plot the correlation matrix
mt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
mt.show()

# Select features with high correlation to the target variable
corr_with_target = corr_matrix['Outcome'].abs().sort_values(ascending=False)
print(corr_with_target)
```

```python
# Separate features and target
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Apply SelectKBest to select top k features
selector = SelectKBest(f_classif, k=5)
X_new = selector.fit_transform(X, y)

# Get the selected features
selected_features = X.columns[selector.get_support()]
print(selected_features)
```

## Dimensionality Reduction

```python
model = RandomForestClassifier()

# Apply RFE
rfe = RFE(model, n_features_to_select=5)
X_rfe = rfe.fit_transform(X, y)

# Get the selected features
selected_features = X.columns[rfe.support_]
print(selected_features)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Step 7: Analyze explained variance
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)


# Step 8: Reduce the dataset to 4 principal components
pca = PCA(n_components=4)
X_pca_reduced = pca.fit_transform(X_scaled)


# Explained variance ratio
print(pca.explained_variance_ratio_)
```

```python
# Convert the reduced dataset to a DataFrame
X_pca_df = pd.DataFrame(X_pca_reduced, columns=[f'PC{i+1}' for i in range(4)])
X_pca_df['Outcome'] = y

# Step 9: Visualize the PCA-transformed data (PC1 vs PC2)
mt.figure(figsize=(8, 6))
for outcome in [0, 1]:
    plt.scatter(
        X_pca_reduced[y == outcome, 0],  # PC1
        X_pca_reduced[y == outcome, 1],  # PC2
        label=f'Outcome {outcome}',
        alpha=0.7
    )

mt.title('PCA Visualization: PC1 vs PC2')
mt.xlabel('Principal Component 1 (PC1)')
mt.ylabel('Principal Component 2 (PC2)')
mt.legend()
mt.grid()
mt.show()
```

```python
# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot the result
mt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
mt.show()
```
